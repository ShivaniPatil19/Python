```python
import pandas as pd

# reads the movie data into a data frame
data=pd.read_csv("movie_metadata.csv")

# prints the first ten rows of the dataset
data.head(10)

```

```python
# gives the statistics for the column imdb_score
print(data['imdb_score'].describe())
```

```python
#prints the unique movie titles
print(data['movie_title'].unique)
```

```python
#displays the movie title the budget and director name 
data[['movie_title','budget','director_name']]
```

```python
#displays the movies with duration more than two hours 
data[data['duration']>120]
```

```python
import numpy as np

#replaces the missing values with NaN
data.replace(np.nan)

#counts the number of null values in each column
print(data.isnull().sum())

data.head(10)
```

```python
#replaces the missing values with the mean of the columns
data.fillna(data.mean(),inplace=True)
data.head(10)
```

```python
dataa=data.dropna(thresh=28)
dataa.isnull().sum()
```

```python
dataa.duplicated().sum() # checks the no of duplicated records
```

```python
dat=dataa.drop_duplicates(keep='first') #drops the duplicated records and keeps the first instance of the row and removes the other duplicated rows
dat.duplicated().sum()
```
